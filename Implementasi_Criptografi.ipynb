# PROSES ENCODING MENGGUNAKAN METODE FERNET

```python
# mendelkarasikan kunci fernet
key = Fernet.generate_key()
 
# Mendeklarasikan Huruf Acak sebagai bahan encoding
with open('filekey.key', 'wb') as filekey:
   filekey.write(key)
```

```python
# Melihat Deklarasikan Huruf Acuan 
key
```

```python
with open('filekey.key', 'rb') as filekey:
    key = filekey.read()
fernet = Fernet(key)
 
# Membuka File yang akan di enkripsi
with open('Status Nutrition Dataset.csv', 'rb') as file:
    original = file.read()
     
# Melakukan Enkripsi
encrypted = fernet.encrypt(original)
 
# Memulai proses enkripsi
with open('Status Nutrition Dataset Encripsi.csv', 'wb') as encrypted_file:
    encrypted_file.write(encrypted)
```

```python
enkripsi_dataset = pd.read_csv('Status Nutrition Dataset Encripsi.csv')
enkripsi_dataset
```

* Mengembalikan  Dataset ke Bentuk Semula

```python
# Mendeklarasikan Kunci Kembali
fernet = Fernet(key)
 
# Membuka Dataset yang sudah di enkripsi
with open('Status Nutrition Dataset Encripsi.csv', 'rb') as enc_file:
    encrypted = enc_file.read()
 
# Melakukan Dekripsi Dataset
decrypted = fernet.decrypt(encrypted)
 
# melakukan dekirpsi dataset
with open('Dataset Status Gizi Dekripsi.csv', 'wb') as dec_file:
    dec_file.write(decrypted)
```

```python
dekripsi_dataset = pd.read_csv('Dataset Status Gizi Dekripsi.csv')
dekripsi_dataset
```
